```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression 
%matplotlib inline
train_data=pd.read_csv(r"C:\Users\temp\Downloads\train.csv")
test_data=pd.read_csv(r"C:\Users\temp\Downloads\test.csv")
```

```python
train_data[['Survived','Age']].corr()
#this state that as age increases survival rate decrease
dummy=pd.get_dummies(train_data['Sex'])
train_data=pd.concat([train_data,dummy],axis=1)

df=train_data.isnull()
for column in df.columns.values.tolist():
    print(df[column].value_counts())
    print("")

train_data=train_data.drop(columns=train_data[['Cabin']])

df['Age'].fillna(df['Age'].median(), inplace=True)
train_data['Embarked'].fillna(train_data['Embarked'].mode()[0], inplace=True)
train_data['male'] = train_data['male'].astype(int)
train_data['female'] = train_data['female'].astype(int)

train_data['is_child'] = train_data['Age'] < 16  # This will be True/False
train_data['is_child'] = train_data['is_child'].astype(int)  # Now 1/0
```

```python
sns.heatmap(train_data.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.show()
```

```python
import plotly.express as px
import plotly.graph_objects as go
fig = px.bar(
    train_data.groupby('Sex')['Survived'].mean().reset_index(),
    x='Sex', y='Survived',
    title='Survival Rate by Sex',
    labels={'Survived': 'Survival Rate', 'Pclass': 'Passenger Class'},
    color='Sex',
    color_discrete_map={
        'male': 'skyblue',
        'female': 'pink'}
)
fig.show()
```

```python
import plotly.express as px

# Create a mapping from number to name
class_names = {1: 'First Class', 2: 'Second Class', 3: 'Third Class'}

# Prepare data
pclass_data = train_data.groupby('Pclass')['Survived'].mean().reset_index()

# Map the class numbers to names
pclass_data['Class Name'] = pclass_data['Pclass'].map(class_names)

# Plot using class names
fig = px.bar(
    pclass_data,
    x='Class Name',
    y='Survived',
    color='Class Name',
    title='Survival Rate by Passenger Class',
    color_discrete_map={
        'First Class': 'gold',
        'Second Class': 'silver',
        'Third Class': 'brown'
    }
)

fig.show()
 
```
